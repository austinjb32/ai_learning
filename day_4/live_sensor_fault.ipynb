```python
import os
import pandas as pd
import numpy as np

folder_path = '1st_test'
files = sorted(os.listdir(folder_path))
```

```python
data = []
for file in files:
    raw = np.loadtxt(os.path.join(folder_path, file))
    # Mean or RMS of each sensor signal per file
    stats = np.mean(raw, axis=0)
    data.append(stats)
```

```python
df = pd.DataFrame(data, columns=["Sensor1", "Sensor2", "Sensor3", "Sensor4", "Sensor5", "Sensor6", "Sensor7", "Sensor8"])
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df_scaled.head()
```

```python
import time

def stream_data(df, delay=1):
    for i in range(len(df)):
        row = df.iloc[i]
        print(f"⏱️ Time: {i} → Sensor values: {row.to_dict()}")
        time.sleep(delay)  # Simulates 1-second interval
```

```python
stream_data(df_scaled, delay=0.5)
```

```python
import streamlit as st
import pandas as pd
i = st.slider("Time Step", 0, len(df)-1)
row = df.iloc[i]
st.line_chart(row)
stream_data(df_scaled, delay=0.5)
```

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(contamination=0.02, random_state=42)
model.fit(df_scaled)
```

```python
def detect_anomaly(model, row):
    pred = model.predict([row])
    return 'FAULT ⚠️' if pred[0] == -1 else 'Normal ✅'

def stream_with_detection(df, model, delay=1):
    for i in range(len(df)):
        row = df.iloc[i]
        status = detect_anomaly(model, row)
        print(f"⏱️ Time: {i} | {status} | Sensor: {row.to_dict()}")
        time.sleep(delay)
        
stream_with_detection(df_scaled, model, delay=0.5)
```

```python
import streamlit as st
import pandas as pd
import time
from sklearn.ensemble import IsolationForest

# Load your processed dataset
model = IsolationForest(contamination=0.02).fit(df_scaled)

st.title("🔧 Real-Time Sensor Fault Detection")

status_placeholder = st.empty()
chart_placeholder = st.line_chart(df.iloc[:1])

for i in range(len(df)):
    row = df.iloc[i]
    prediction = model.predict([row])[0]
    label = "⚠️ FAULT DETECTED" if prediction == -1 else "✅ Normal"
    
    status_placeholder.subheader(f"⏱ Time: {i} | Status: {label}")
    chart_placeholder.add_rows(df.iloc[[i]])

    time.sleep(0.5)  # Simulates real-time
```

```python
results = []

for i in range(len(df)):
    row = df.iloc[i]
    prediction = model.predict([row])[0]
    status = "Fault" if prediction == -1 else "Normal"

    results.append({
        "timestamp": i,
        "status": status,
        **row.to_dict()
    })

pd.DataFrame(results).to_csv("fault_detection_log.csv", index=False)
st.success("Log saved as fault_detection_log.csv")
```
